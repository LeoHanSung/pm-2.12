# 딕셔너리 만들기

```python
# 딕셔너리 생성 위해 미리 제작한 엑셀 파일 불러오기
import pandas as pd
stat = pd.read_excel('../data/station.xlsx')
stat
```

```python
# 딕셔너리1 만들기 (한글 동이름 - 영어 동이름)
aa1 = list(stat['dong'])
aa2 = list(stat['name'])
dic = dict(zip(aa1,aa2))
dic
```

```python
# 딕셔너리 2 만들기 (영어 동이름 - 한글 동이름)
dicr = dict(zip(aa2,aa1))
dicr
```

```python
# 딕셔너리 3 만들기 (영어 동이름 - 스테이션 이름)
aa2 = list(stat['name'])
aa3 = list(stat['station'])
dic2 = dict(zip(aa3,aa2))
dic2
```

```python
# 딕셔너리 4 만들기 (영어 동이름 - 행정동코드)
aa2 = list(stat['name'])
aa4 = list(stat['code'])
dic3 = dict(zip(aa4,aa2))
dic3
```

# 데이터 합치기

```python
# 유동인구 데이터 불러오기
popul = pd.read_csv('../data/floatpopul.csv')
popul
```

```python
# 유통 매출지수 데이터 불러오기
retail = pd.read_excel('../data/tretail.xlsx')
retail
```

```python
# 미세먼지 데이터 불러오기 (이전 작업을 통해 뉴스기사 개수 데이터도 포함되었음)

import glob
import os
import pandas as pd

os.getcwd()

filenames = glob.glob('../data/weatherdata/*.csv')
```

```python
# 미세먼지 데이터 파일명을 리스트로
filenames
```

```python
for i in range(len(filenames)) :
    f = filenames[i]
    
    # 파일명에서 스테이션 이름 추출
    fsta = f[20:-9]
    
    # 딕셔너리로 영어 동이름 가져오기
    en = dic2[fsta]
    readf = pd.read_csv(f)
    
    # 추후 merge 작업 위해 날짜 데이터 타입 str으로 변경
    dlis = list(readf['Date'])
    dlis = [str(d) for d in dlis]
    readf['Date'] = dlis
    readf
    
    # 영어 동이름을 key로 설정, 딕셔너리를 통해 한글 동이름 가져오기
    kn = dicr[en]

    # 한글 동이름에 해당하는 유통 매출지수 데이터 가져오기 
    retailp = retail[retail['ADMD_NM'] == kn]
    
    # 행-열 변환
    retailp = retailp.transpose()
    
    # 유통 매출지수 데이터 행 이름을 리스트로
    pcol = list(retailp.columns)
    
    # 유통 매출지수 데이터의 인덱스를 새로운 행으로 설정 (날짜 행)
    tl = list(retailp.index)
    retailp['Date'] = tl
    
    # 유통 매출지수 데이터 인덱스 초기화 및 컬럼명 변경, 필요없는 값 제거
    retailp = retailp.reset_index(drop=True)
    retailp = retailp.rename(columns = {pcol[0]:'retail'})
    retailp = retailp.drop(0)
    
    # 영어 동이름에 해당하는 유동인구 데이터 가져오기
    popul2 = popul[popul['dong']==en]
    
    # 유동인구 데이터 인덱스 초기화 및 컬럼명 변경
    popul2 = popul2.rename(columns = {'STD_YMD':'Date'})
    popul2 = popul2.reset_index(drop=True)
    
    # 추후 merge 작업 위해 날짜 데이터 타입 str으로 변경
    fl = list(popul2['Date'])
    fl = [str(k) for k in fl]
    popul2 ['Date'] = fl
   
    # 날짜 기준으로 유통 매출지수 데이터, 유동인구 데이터 merge
    re1 = retailp.merge(popul2, on='Date', how='outer')
    
    # 날짜 기준으로 위의 데이터, 미세먼지 데이터 merge
    fin1 = re1.merge(readf, on='Date', how='outer')

    # 위의 데이터 결측치 제거, 인덱스 초기화, 필요한 열만 가져오기
    fin1 = fin1.dropna()
    fin1 = fin1.reset_index(drop=True)
    fin1 = fin1[['Date','pm10','pm25','COUNT_T','TMST','retail']]
    
    # 소수점 세 자리까지 자르는 작업
    w1 = list(fin1['retail'])
    w1 = [round(w,3) for w in w1]
    fin1['retail'] = w1
    v1 = list(fin1['pm10'])
    v1 = [round(v,3) for v in v1]
    fin1['pm10'] = v1
    q1 = list(fin1['pm25'])
    q1 = [round(q,3) for q in q1]
    fin1['pm25'] = q1
    u1 = list(fin1['TMST'])
    u1 = [round(u,3) for u in u1]
    fin1['TMST'] = u1
    
    # 각각 새로운 파일로 저장
    
    if i < 9 :      
        fin1.to_csv('../data/total/'+str(0)+str(i+1)+'_'+en+'.csv',index=False)
        
    else :
        fin1.to_csv('../data/total/'+str(i+1)+'_'+en+'.csv',index=False)
    
    print(str(i+1)+'번째 완료되었습니다')
    print('---------------------------------------')
```
