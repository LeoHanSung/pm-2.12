# 딕셔너리 생성

```python
# 딕셔너리 생성 위해 미리 제작한 엑셀 파일 불러오기
import pandas as pd
stat = pd.read_excel('../data/station.xlsx')
stat
```

```python
# 딕셔너리 만들기 (영어 동이름 - 한글 동이름)
aa1 = list(stat['name'])
aa2 = list(stat['dong'])
dic = dict(zip(aa1,aa2))
dic
```

# 검색률 합치기

```python
import glob
import os

os.getcwd()

filenames = glob.glob('../data/total/*.csv')
```

```python
filenames
```

```python
i = 0

for f in filenames :
    
    # 각각의 파일 불러오기
    total = pd.read_csv(f)
    
    # 날짜 데이터 타입 str으로
    lis = list(total['Date'])
    lis = [str(l) for l in lis]
    total['Date'] = lis
    
    # 사전 조사한 동별 인구비율 파일 가져오기
    poprate = pd.read_excel('../data/dongpoprate.xlsx')
    
    # 영어 동이름으로 한글 동이름 출력
    kn = dic[f[17:-4]]
    
    # 한글 동이름에 맞는 동별 인구비율 가져오기
    poprate1 = poprate[poprate['dong']==kn]
    
    # 인덱스를 리스트로 -> 추후 대입
    poin = list(poprate1.index)
    
    # 사전에 조사한 구글 트렌드 미세먼지 검색률 (서울특별시) 파일 가져오기
    search = pd.read_csv('../data/seoulsearchrate.csv')
    
    # 날짜 데이터 타입 str으로 바꾸고 형식 맞추기
    dlist = list(search['Date'])
    dlist = [str(b)[:4]+str(b)[5:7]+str(b)[8:10] for b in dlist]
    search['Date'] = dlist
    
    # 검색률 * 해당 동별 인구비율
    srate = search['search'] * poprate['rate'][poin[0]]
    search['search'] = srate
    
    # 원래 파일과 검색률 합치는 작업
    fin = search.merge(total, on = 'Date', how = 'outer')
    fin = fin.dropna()
    fin = fin.reset_index(drop=True)
    
    # 각 열의 값들 소수점 정리
    
    w10 = list(fin['retail'])
    w10 = [round(w,3) for w in w10]
    fin['retail'] = w10

    v10 = list(fin['pm10'])
    v10 = [round(v,3) for v in v10]
    fin['pm10'] = v10

    q10 = list(fin['pm25'])
    q10 = [round(q,3) for q in q10]
    fin['pm25'] = q10

    u10 = list(fin['TMST'])
    u10 = [round(u,3) for u in u10]
    fin['TMST'] = u10

    k10 = list(fin['search'])
    k10 = [round(k,5) for k in k10]
    fin['search'] = k10

    fin.to_csv('../data/total2/'+f[14:-4]+'.csv',index=False)
    
    i += 1
    print(str(i)+'번째 완료되었습니다.')
    print('=================================')
```
