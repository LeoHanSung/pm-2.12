```python
import pandas as pd
deep = pd.read_csv('../data/rawdataNew.csv')
raw = pd.read_csv('../data/rawdata.csv')
```

```python
# 딥인덱스 포함한 종합 데이터 가져오기
deep
```

```python
# 딥러닝에 사용된 원본 데이터 가져오기
raw
```

```python
# 원본 데이터 날짜 가져와 열로 추가 
deep['Date'] = raw['Date']
```

```python
deep
```

```python
# 날짜 기준으로 각 열 평균 내기
mean_d = deep.pivot_table(values=['pm10','pm25','COUNT_T','TMST','search','retail','Deepindex'],index = 'Date', aggfunc='mean')
mean_d
```

```python
# 인덱스의 날짜 데이터 가져와서 리스트로 만들고 열로 추가
mean_d['Date'] = list(mean_d.index)
dlist = list(mean_d['Date'])
dd = [str(d) for d in dlist]
mean_d['Date'] = dd

# 인덱스 초기화
mean_d = mean_d.reset_index(drop=True)
```

```python
mean_d
```

```python
# 유통 카테고리별 비중 데이터 가져오기 
category = pd.read_excel('../data/catretail.xlsx')

# NaN값 0으로 채우기
category = category.fillna(0)
category
```

```python
# 필요없는 열 제거
del category['BOR_CD']
del category['ADMD_CD']

# 행 이름 변경
category = category.rename(columns = {'OPER_DT': 'Date'})
category
```

```python
# 각 카테고리별 비중에 매출지수 곱하고 열값 변경하기

category['LCLS_10_P'] = category['AMT_IND'] * category['LCLS_10_P']
category['LCLS_20_P'] = category['AMT_IND'] * category['LCLS_20_P']
category['LCLS_30_P'] = category['AMT_IND'] * category['LCLS_30_P']
category['LCLS_40_P'] = category['AMT_IND'] * category['LCLS_40_P']
category['LCLS_50_P'] = category['AMT_IND'] * category['LCLS_50_P']
category['LCLS_60_P'] = category['AMT_IND'] * category['LCLS_60_P']
category['LCLS_70_P'] = category['AMT_IND'] * category['LCLS_70_P']
category['LCLS_80_P'] = category['AMT_IND'] * category['LCLS_80_P']

category
```

```python
# 날짜 기준으로 각 열 평균 내기
mean_c = category.pivot_table(values=['AMT_IND','LCLS_10_P','LCLS_20_P','LCLS_30_P','LCLS_40_P','LCLS_50_P','LCLS_60_P','LCLS_70_P','LCLS_80_P'],index = 'Date', aggfunc='mean')
mean_c
```

```python
ilist=list(mean_c.index)
ii = [str(i) for i in ilist]
mean_c['Date']=ii
mean_c
```

```python
# 인덱스 초기화
mean_c = mean_c.reset_index(drop=True)
mean_c
```

```python
results = mean_c.merge(mean_d, on='Date', how='outer')
results 
```

```python
# Null값 제거
results = results.dropna()
results
```

```python
results2 = results[['Date','Deepindex','AMT_IND','LCLS_10_P','LCLS_20_P','LCLS_30_P','LCLS_40_P','LCLS_50_P','LCLS_60_P','LCLS_70_P','LCLS_80_P']]
results2
```

```python
results2.to_csv('../data/Deepcategory.csv',index=False)
```
