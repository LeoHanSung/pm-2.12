```python
# 시간별 유동인구 파일 불러오기

import glob
import os
import pandas as pd

os.getcwd()

filenames = glob.glob('../data/popul/*.csv')
```

```python
# 파일명을 리스트로
filenames
```

```python
# 각 파일의 행 개수 확인
for f in filenames :
    aa = pd.read_csv(f)
    print(len(aa))
```

```python
# 각 파일들을 하나의 파일로 concat (행으로 누적됨)
for i in range(len(filenames)) :
    if i == 0 :
        fp1 = pd.read_csv(filenames[0])
    elif i == 1 :
        fp2 = pd.read_csv(filenames[1])
        result = pd.concat([fp1,fp2])
    else :
        fp3 = pd.read_csv(filenames[i])
        result = pd.concat([result,fp3])
        
    print(str(i+1)+'번째 완료했습니다')
```

```python
# 유동인구 일자별 평균 구하기
result['TMST'] = (result['TMST_00']+result['TMST_01']+result['TMST_02']+result['TMST_03']+result['TMST_04']+result['TMST_05']
                  +result['TMST_06']+result['TMST_07']+result['TMST_08']+result['TMST_09']+result['TMST_10']+result['TMST_11']+
                 +result['TMST_12']+result['TMST_13']+result['TMST_14']+result['TMST_15']+result['TMST_16']+result['TMST_17']+
                 +result['TMST_18']+result['TMST_19']+result['TMST_20']+result['TMST_21']+result['TMST_22']+result['TMST_23'])/24
```

```python
# 필요한 열만 불러오기 (날짜, 동이름, 유동인구)
result2 = result[['STD_YMD', 'HDONG_NM', 'TMST']]
result2
```

```python
# 소수점 세 자리까지 자르기
result2['TMST'] = round(result2['TMST'],3)
```

```python
# 동 이름 리스트로 만들기
hdong = list(result2['HDONG_NM'])
hdong
```

```python
# 미리 엑셀로 작업한 파일 불러오기
import pandas as pd
stat = pd.read_excel('../data/station.xlsx')
stat
```

```python
# 딕셔너리 만들기 (한글 동이름 - 영어 동이름)
aa1 = list(stat['dong'])
aa2 = list(stat['name'])
dic = dict(zip(aa1,aa2))
dic
```

```python
# 딕셔너리에 한글 동이름을 key로 대입해서 영어 동이름을 value로 뽑기
ename = [dic[h] for h in hdong]
ename
```

```python
# 영어 동이름 리스트를 새로운 열로 생성
result2['dong'] = ename
result2
```

```python
# 기존 한글 동이름 열 제거 
del result2['HDONG_NM']
```

```python
# 열 순서 변경
result2 = result2[['STD_YMD','dong','TMST']]
```

```python
# 유동인구 정제 데이터 새로운 파일로 저장
result2.to_csv('../data/floatpopul.csv', index = False)
```
